```python
import pandas as pd
```

```python
data = pd.read_csv('adult.data.csv')
data.head(3)
```

```python
#race count
race_count = data.groupby('race', as_index=True).agg({'race':'count'})
race_count
```

```python
race_count_1 = data['race'].value_counts().tolist()
race_count_1
```

```python
average_age_men = data[data['sex'] == 'Male']['age'].mean()
average_age_men
```

```python
   # What is the percentage of people who have a Bachelor's degree?
education_seg = data['education'].value_counts()
print(education_seg['Bachelors'])
```

```python
percentage_bachelors = data['education'].value_counts()['Bachelors']/data['education'].value_counts().sum() * 100
percentage_bachelors
```

```python
#percentage with higher education
higher_education = data[data['education'].isin(['Masters', 'Bachelors', 'Doctorate'])]
print(len(higher_education['education'])/len(data['education'])*100)
#with lower education
lower_education = data[~data['education'].isin(['Masters', 'Bachelors', 'Doctorate'])]
print(len(lower_education['education'])/len(data['education'])*100)
```

```python
higher_50k = higher_education[higher_education['salary'].str.startswith('>50K')]
percentatge_higher_above_50k = len(higher_50k['salary'])/len(data['salary']) *100
```

```python
lower_50k = lower_education[lower_education['salary'].str.startswith('>50K')]
percentatge_lower_above_50k = len(lower_50k['salary'])/len(data['salary']) *100
print(percentatge_lower_above_50k)
```

```python
 # What is the minimum number of hours a person works per week (hours-per-week feature)?
minimum_hrs = data['hours-per-week'].min()
```

```python
 # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = data[data['hours-per-week']==minimum_hrs]
len(num_min_workers[num_min_workers['salary'].str.startswith('>50K')]['salary'])
```

```python
with_50k = data[data['salary'].str.startswith('>50K')]
country_earn_counts = with_50k.groupby('native-country').size().reset_index(name='count_over_50k')
country_earn_counts.sort_values(by='count_over_50k', ascending=False).iloc[0]['native-country']
# with_50k
```

```python
  # What country has the highest percentage of people that earn >50K?
with_50k = data[data['salary'].str.startswith('>50K')]
country_earn_counts = with_50k.groupby('native-country').size().reset_index(name='count_over_50k')
highest_earning_country = country_earn_counts.sort_values(by='count_over_50k', ascending=False).iloc[0]['native-country']
highest_earning_country_percentage = round(country_earn_counts.sort_values(by='count_over_50k', ascending=False)['count_over_50k'].max()/len(with_50k['salary'])*100,1)

print(highest_earning_country,highest_earning_country_percentage)
print(country_earn_counts)
```

```python
gr_country = data.groupby('native-country', as_index=False).agg({
    'salary': ['count', lambda x: (x.str.startswith('>50K')).sum()]
})
# Rename columns for better readability
gr_country.columns = ['country', 'total_count', 'over_50k_count']
# Calculate the ratio of salaries over 50K to total salaries
gr_country['ratio'] = round(gr_country['over_50k_count'] / gr_country['total_count']*100,1)

# Sort by country name
gr_country = gr_country.sort_index() 
cntry = gr_country[gr_country['ratio'] ==gr_country['ratio'].max()]['country']
print(cntry.tolist()[0])
print(gr_country['ratio'].max())
```
